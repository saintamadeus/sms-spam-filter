# Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# 1. Dataset Loading
df = pd.read_csv('../data/spam.csv', encoding='latin-1')
df = df[['v1', 'v2']]
df.columns = ['target', 'text']
df.head()
```

```python
# 3. Class Distribution
df['target'].value_counts()
```

```python
# 4. Visualization
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='target')
plt.title('Spam vs Ham')
plt.show()
```

```python
df['length'] = df['text'].apply(len)
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='length', hue='target', bins=50)
plt.title('Message Length Distribution')
plt.show()
```

```python
# 5. Message Statistics
df.groupby('target')['length'].describe()
```

```python
# 6. Word Clouds
spam_words = ' '.join(list(df[df['target'] == 'spam']['text']))
spam_wc = WordCloud(width=500, height=500, background_color='white').generate(spam_words)

plt.figure(figsize=(10, 8))
plt.imshow(spam_wc)
plt.axis('off')
plt.title('Spam Word Cloud')
plt.show()
```
